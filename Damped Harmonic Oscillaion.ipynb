## Damped Harmonic Oscillaion

```python
import matplotlib.pyplot as plt

def f1(x,y,z):
    return z
def f2(x,y,z,b,w):
    return -2*b*z-w**2*y

def D(xi,xf,y,z,n,w,b):
    X=[]
    Y=[]
    h=(xf-xi)/(n-1)
    for i in range(n):
        x=xi+i*h
        X.append(x)
        z=z+h*f2(x,y,z,b,w)
        y=y+h*f1(x,y,z)
        Y.append(y)
    return X,Y

z1,z2=D(0,10,1,1,500,6,18)#Overdamped Case
z3,z4=D(0,10,1,1,500,6,6)#Critically Damped Case
z5,z6=D(0,10,1,1,500,6,1)#Underdamped Case
    
plt.plot(z1,z2,c='r')
plt.plot(z3,z4,c='g')
plt.plot(z5,z6,c='b')
plt.xlabel('Time',size=16)
plt.ylabel('Amplitude',size=16)
plt.axhline(linestyle = '--')
plt.show()
```
