## Euler's Method

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return -y

x=0
y=1
n=20

xx=[]
yy=[]

xi=0
xf=5
h=(xf-xi)/(n-1)

for i in range (n):
    x=xi+i*h
    xx.append(x)
    y=y+h*f(x,y)
    yy.append(y)
    
a=0
b=5
g=(b-a)/(n-1)

aa=[]
bb=[]

for i in range (n):
    a=a+i*g
    aa.append(a)
    b=np.exp(-a)
    bb.append(b)
    
plt.plot(xx,yy)
plt.plot(aa,bb)
plt.show()
```
