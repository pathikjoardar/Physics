```python
"""f(x)=1 for (-pi<x<0)
       =0 for (0<x<pi)"""

import numpy as np
import matplotlib.pyplot as plt

def Trap(xi,xf,n,f):
    x=np.linspace(xi,xf,n)
    h=(xf-xi)/(n-1)
    y=f(x)
    sum=0
    for i in range(1,len(y)-1):
        sum += y[i]
    sum = 0.5*h*(y[0]+2*sum+y[len(y)-1])
    return(sum)


def a(n):
    if n == 0:
        return Trap(-np.pi,0,1000,f=lambda x:x**0)/np.pi
    else:
        return Trap(-np.pi,0,1000,f=lambda x:np.cos(n*x))/np.pi

def b(n):
    return Trap(-np.pi,0,1000,f=lambda x:np.sin(n*x))/np.pi


def Fourier(x,n):
    fourier = a(0)/2
    for i in range(1,n+1):
        fourier += a(i)*np.cos(i*x)+b(i)*np.sin(i*x)
    return (fourier)

x=np.linspace(-3*np.pi,3*np.pi,1000)    
y1= Fourier(x,10)
y2= Fourier(x,20)
y3= Fourier(x,40)

plt.figure(figsize=(16, 8))
plt.plot(x,y1)
plt.plot(x,y2,"--")
plt.plot(x,y3,"")
plt.show()
```
