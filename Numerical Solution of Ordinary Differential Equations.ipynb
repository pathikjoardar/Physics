## Euler's Method

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return -y
def True_Solution(x1):
    return np.exp(-x1)

x = 0
y = 1
n = 500
h = 0.03

def Euler(f, x, y):
    X = []
    Y = []
    for i in range(n):
        x = x + i * h
        X.append(x)
        y = y + h * f(x, y)
        Y.append(y)
    return X, Y

X, Y = Euler(f, x, y)

x1=0

A=[]
B=[]

for i in range (n):
    x1=x1+i*h
    A.append(x1)
    b=True_Solution(x1)
    B.append(b)
    
plt.plot(X,Y,'b')
plt.show()
plt.plot(A,B,'g')
plt.show()
```

## Modified Euler Method

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return -y
def True_Solution(z):
    return np.exp(-z)

x0=0
y0=1
h=0.04
n=400
X=[]
Y=[]

def Mod_Euler(f ,x ,y ):
    y = y0 + h*f(x,y)
    for i in range(n):
        x = x0 + i*h
        X.append(x)
        y = y0 + 0.5*h*(f(x0,y0) + f(x,y))
        Y.append(y)
    return X,Y

X, Y=Mod_Euler(f,x,y)

z=0

A=[]
B=[]

for i in range (n):
    z = z + i*h
    A.append(z)
    b = True_Solution(z)
    B.append(b)
    
plt.plot(A,B,'g')
plt.show()
plt.plot(X,Y,'b')
plt.show()
```

## Runge-Kutta 2nd Order (RK2)

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return -y

def True_Solution(x):
    return np.exp(-x)

x = 0
y = 1
n = 400
h = 0.04
X = []
Y = []

def RK2(f, x, y, n, h):
    for i in range(n):
        k1 = h * f(x, y)
        k2 = h * f(x + h, y + k1)
        y = y + 0.5 * (k1 + k2)
        Y.append(y)
        x = x + h
        X.append(x)
    return X, Y

X, Y = RK2(f, x, y, n, h)

x1 = 0
A = []
B = []

for i in range(n):
    x1 = x1 + h
    A.append(x1)
    b = True_Solution(x1)
    B.append(b)
    
plt.xlabel('x')
plt.ylabel('y')
plt.title('RK2 Method for y\' = -y')
plt.plot(X, Y, 'b', label='RK2')
plt.legend()
plt.show()
plt.plot(A, B, 'g', label='True Solution')
plt.legend()
plt.show()
```

## Runge-Kutta 4th Order (RK4)

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return -y

def True_Solution(x):
    return np.exp(-x)

x = 0
y = 1
n = 400
h = 0.04
X = []
Y = []

def RK4(f, x, y, n, h):
    for i in range(n):
        k1 = h * f(x, y)
        k2 = h * f(x + 0.5 * h, y + 0.5 * k1)
        k3 = h * f(x + 0.5 * h, y + 0.5 * k2)
        k4 = h * f(x + h, y + k3)
        y = y + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        Y.append(y)
        x = x + h
        X.append(x)
    return X, Y

X, Y = RK4(f, x, y, n, h)

x1 = 0
A = []
B = []

for i in range(n):
    x1 = x1 + h
    A.append(x1)
    b = True_Solution(x1)
    B.append(b)

plt.xlabel('x')
plt.ylabel('y')
plt.title('RK4 Method for y\' = -y')
plt.plot(X, Y, 'b', label='RK4')
plt.legend()
plt.show()
plt.plot(A, B, 'g', label='True Solution')
plt.legend()
plt.show()
```
