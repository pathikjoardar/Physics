## Simulation of coin tossing

```python
import numpy as np
import matplotlib.pyplot as plt

n=100000
N=range(1,100000)
X=np.random.choice([0,1],size=n)
Y=[np.sum(X[:i])/i for i in N]

plt.plot(N,Y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random

N=range(1,10000)
def Coin_Tossing(N):
    h=0
    H=[]
    for i in N:
        h=h+random.choice([0,1])
        H.append(h/i)
    return H
H=Coin_Tossing(N)

plt.plot(N, H)
plt.xlabel('Sample Size (N)')
plt.ylabel('Running Average (H)')
plt.title('Running Average of Coin Tossing')
plt.show()
```

## Weldon's Dice Experiment

```python
import random
import matplotlib.pyplot as plt
import numpy as np

num_trials = 26306
num_dice = 12
face_to_count = 6
weldons_results = []

for i in range(num_trials):
    rolls = [random.randint(1, 6) for _ in range(num_dice)]
    num_face_to_count = rolls.count(face_to_count)
    weldons_results.append(num_face_to_count)

n = num_dice
p = 1/6

def binom_dist(n, k, p):  
    def fac(a):
        if a <= 1:
            return 1
        else: 
            return a*fac(a-1)
        
    def C(m, r): return fac(m)/(fac(r)*fac(m-r))
    
    return C(n, k)*(p**k)*(1-p)**(n-k)

binom_probs = [binom_dist(n, k, p) for k in range(n+1)]

plt.bar(range(n+1), binom_probs, label='Binomial Distribution')
plt.hist(weldons_results, bins=range(n+2), density=True, alpha=0.5, label="Weldon's Dice Experiment")
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title("Weldon's Dice Experiment vs Binomial Distribution")
plt.legend()
plt.show()
```
