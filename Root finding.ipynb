## Finding the roots using Newton Raphson method

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    y=(x**3)-5*(x**2)+3*(x)
    return y
def g(x):
    y=3*(x**2)-10*(x)+3
    return y
x=2
err=0.0001
h=f(x)/g(x)

while abs(f(x))>err:
    h=f(x)/g(x)
    x=x-h
    
print(x)

xx=np.linspace(-10,10,500)
yy=f(xx)

plt.plot(xx,yy,'-',label='Graph of Function')
plt.title("The Graph of the function and its root")
plt.text(x,0,f"The root is at({x:.2f})")
plt.plot(x,0,'ro')
plt.xlabel('Value of x')
plt.ylabel('value of f(x)')
plt.axhline()
plt.grid()
plt.show()
```

## Finding roots using Bisection method

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    y=(x**3)-5*(x**2)+16
    return y
    
a=-5
b=5
err=0.0001

while f(a)*f(b)>0:
    print('No roots in this interval')
    break
    
while abs(b-a)>=err:
    xm=(a+b)/2
    
    if f(xm)==0:
        print('Root =',xm)
        break
    
    if f(a)*f(xm)<0:
        b=xm
        
    else:
        a=xm
        
print('Root =',(a+b)/2)

xx=np.linspace(-10,10,500)
yy=f(xx)

plt.plot(xx,yy,'-',label='Graph of Function')
plt.title("The Graph of the function and its root")
plt.text((a+b)/2,0,f"The root is at({(a+b)/2:.2f})")
plt.plot((a+b)/2,0,'ro')
plt.xlabel('Value of x')
plt.ylabel('value of f(x)')
plt.axhline()
plt.grid()
plt.show()
```
