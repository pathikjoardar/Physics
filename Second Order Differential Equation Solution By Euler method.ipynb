## Second Order Differential Equation Solution By Euler method

d2y/dx2+lam*(dy/dx)+k*y=0

Let, dy/dx=z 

```python
import matplotlib.pyplot as plt

X=[]
Y=[]
def f1(x,y,z):
    return z
def f2(x,y,z):
    return -lam*z-k*y

h=0.1
lam=0.5
k=2
z,y,x=1,0,0
for i in range(200):
    z=z+h*f2(x,y,z)
    y=y+h*f1(x,y,z)
    x=x+h
    X.append(x)
    Y.append(y)
    
plt.plot(X,Y)
plt.show()
    
```
