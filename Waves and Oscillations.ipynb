## Damped Oscillation

```python
import matplotlib.pyplot as plt

def f1(t,x,z):
    return z
def f2(t,x,z,b,w):
    return -2*b*z-w**2*x

def D(ti,tf,x,z,n,w,b):
    T=[]
    X=[]
    h=(tf-ti)/(n-1)
    for i in range(n):
        t=ti+i*h
        T.append(t)
        z=z+h*f2(t,x,z,b,w)
        x=x+h*f1(t,x,z)
        X.append(x)
    return T,X

z1,z2=D(0,10,1,1,500,6,18)#Overdamped Case
z3,z4=D(0,10,1,1,500,6,6)#Critically Damped Case
z5,z6=D(0,10,1,1,500,6,1)#Underdamped Case
    
plt.plot(z1,z2,c='r')
plt.plot(z3,z4,c='g')
plt.plot(z5,z6,c='b')
plt.xlabel('Time',size=16)
plt.ylabel('Amplitude',size=16)
plt.axhline(linestyle = '--')
plt.show()
```

## Forced Oscillation

```python
import matplotlib.pyplot as plt
import numpy as np

def F(ti,tf, n,yi,zi,w,b,f0,p):
    T=[ti]; Y=[yi]
    h=(tf-ti)/(n-1)
    
    for i in range(n-1):
        zi=zi+h*(f0*np.sin(p*ti)-2*b*zi-w**2*yi)
        yi= yi + h*zi
        ti = ti + h
        
        T.append(ti)
        Y.append(zi)
        
    return T, Y
    
n=10000

A,B=F(0,20, n, 1,0,5,0.5,2, 3)

plt.plot(A, B)
plt.axhline(linestyle='--')
plt.show()
```

## Resonance Curve

```python
import matplotlib.pyplot as plt
import numpy as np

def F(ti,tf, n,yi,zi,w,b,f0,p):
    T=[ti]; Y=[yi]
    h=(tf-ti)/(n-1)
    
    for i in range(n-1):
        zi=zi+h*(f0*np.sin(p*ti)-2*b*zi-w**2*yi)
        yi= yi + h*zi
        ti = ti + h
        
        T.append(ti)
        Y.append(zi)
        
    return T, Y
    
n=400
Freq=np.linspace(0.5,5,1000)
A=[]
A1=[]

for p in Freq:
    X1,X2=F(0,20, n, 1,0,5,0.5,2, p)
    y=X2
    amp=np.max(y[-400:])
    A.append(amp)
    
plt.plot(Freq,A)
plt.show()
    
```
